```python
import csv
import math
import statistics
from IPython.display import clear_output
from statistics import mean,stdev
import datetime
from datetime import datetime

checking_files = True

def correlation(list_1,list_2):
    zipped_list = list(zip(list_1,list_2))
    sums = []

    for a,b in zipped_list:
        sums.append(((a-mean(list_1))/stdev(list_1)) * ((b-mean(list_2))/stdev(list_2)))

    r = round(sum(sums)/(len(zipped_list)-1),3)

    return r

while checking_files == True:
    data_location = input("Paste full file path here for .csv file:")
    data = open(data_location)
    csv_data = csv.reader(data)
    data_lines = list(csv_data)
    headers = data_lines.pop(0)
    
    date_col_check = input("Does your data contain a date only column? (Y/N):")
    
    if date_col_check == "Y":
        date_col_value = int(input("Which column? (type single integer value, remember python indexes at 0):"))
        
        start_date = datetime.strptime(data_lines[0][date_col_value], '%m/%d/%y')
        
        for x in range(1,len(data_lines)):
            
            possible_date = datetime.strptime(data_lines[x][date_col_value], '%m/%d/%y')
    
            date_diff = start_date - possible_date
    
            if date_diff.days > 0:
                start_date = datetime.strptime(data_lines[x][date_col_value], '%m/%d/%y')
            else:
                pass
        
        for x in range(0,len(data_lines)):
            data_lines[x][date_col_value] = datetime.strptime(data_lines[x][date_col_value], '%m/%d/%y') - start_date
            data_lines[x][date_col_value] = f'{data_lines[x][date_col_value].days}'
    
    for x in range(0,len(data_lines)):
        for y in range(0,len(data_lines[x])):
            if data_lines[x][y].isdigit() == True:
                data_lines[x][y] = float(data_lines[x][y])
            else:
                pass
    
    pair_list = []
    for x in range(0,len(data_lines[0])):
        pair_list.append(x)
    
    
    list_pairings = []
    for a in range(1,len(pair_list)):
        for x in range(0,len(pair_list)-a):
            list_pairings.append((x,x+a))
    
    
    correlations = []

    for (a,b) in list_pairings:
    
        lista = []
        listb = []
    
        for x in range(0,len(data_lines)):
            lista.append(data_lines[x][a])
            listb.append(data_lines[x][b])
    
        correlations.append([headers[a],headers[b],correlation(lista,listb)])
    
    for x in correlations:
        if abs(x[2]) < 0.2:
            x.append('Very Weak')
        elif abs(x[2]) >= 0.2 and abs(x[2]) < 0.4:
            x.append('Weak')
        elif abs(x[2]) >= 0.4 and abs(x[2]) < 0.6:
            x.append('Moderate')
        elif abs(x[2]) >= 0.6 and abs(x[2]) < 0.8:
            x.append('Strong')
        elif abs(x[2]) >= 0.8 and abs(x[2]) <= 1.0:
            x.append('Very Strong')
            
    for x in correlations:
        print(f'{x[0]} and {x[1]} have an r value of {x[2]} which indicates the correlation is {x[3]}')

    run_again = input('Would you like to import a new file? (Y/N):')
    
    if run_again == 'Y':
        clear_output()
        checking_files = True
    else:
        checking_files = False
```
